## 10. Create 2D list to DataFrame

lists = [[1, 'aaa', 22],
         [2, 'bbb', 25],
         [3, 'ccc', 24]]

```python
lists = [[1, 'aaa', 22], [2, 'bbb', 25], [3, 'ccc', 24]]
```

```python
import pandas as pd 
lists = [[1, 'aaa', 22], [2, 'bbb', 25], [3, 'ccc', 24]]
print(type(lists))
for b in lists:
  for j in b:
    print(j)
dt=zip(lists)
df=pd.DataFrame(dt,columns=["d"])
print(type(df))
print(df)    
```
